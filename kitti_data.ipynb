```python
import numpy as np
import cv2
```

```python
class LabeledObject:
    def __init__(self,label,img_idx,idx):
        sp = label.split()
        self.type = sp[0]               # 'Car', 'Pedestrian', ...
        self.truncation = float(sp[1])  # truncated pixel ratio ([0..1])
        self.occlusion = int(sp[2])     # 0 = visible, 1 = partly occluded, 2 = fully occluded, 3 = unknown
        others = [float(i) for i in sp[3:]]# object observation angle ([-pi..pi])
        self.alpha = others[0]
        # extract 2D bounding box in 0-based coordinates
        self.box2D = others[1:5]    #(xmin,ymin,xmax,ymax)
        # extract 3D bounding box information
        self.h = others[5]     # box width
        self.w = others[6]     # box height
        self.l = others[7]     # box length
        self.t = others[8:11]  # location (x,y,z)
        self.ry = others[11]   # yaw angle
        self.objID = '%06d' % img_idx+'_'+'%02d' % idx
    def __str__(self):
        s = []
        s.append('id:%s' % self.objID)
        s.append('type:%s' % self.type)
        s.append('truncation:%f' % self.truncation)
        s.append('occlusion:%d' % self.occlusion)
        s.append('alpha:%f' % self.alpha)
        s.append('box2D:%s' % self.box2D)
        s.append('h:%f' % self.h)
        s.append('w:%f' % self.w)
        s.append('l:%f' % self.l)
        s.append('t:%s' % self.t)
        s.append('ry:%f' % self.ry)
        return '\n'.join(s)
```

```python
def readLabels(label_dir,img_idx):
    f = open('%s/%06d.txt' % (label_dir,img_idx))
    objects = []
    i = 0
    for l in f:
        objects.append(LabeledObject(l,img_idx,i))
        i += 1
    f.close()
    return objects
```

```python
def readCalibration(calib_dir,img_idx,cam):
    f = open('%s/%06d.txt' % (calib_dir,img_idx))
    i = 0
    for l in f:
        if i==cam:
            sp = l.split()
            p = np.fromstring(' '.join(sp[1:]),sep=' ')
        i = i+1
    f.close()
    p.reshape(4,3).transpose()
    return p
```

```python
def projectToImage(pts_3D, P):
    # project in image
    pts_2D = np.dot(P,np.concatenate((pts_3D,np.ones((1,pts_3D.shape[1])))));
    # scale projected points
    pts_2D[0] = pts_2D[0]/pts_2D[2];
    pts_2D[1] = pts_2D[1]/pts_2D[2];
    pts_2D = pts_2D[0:2]
    return pts_2D
def computeOrientation3D(obj,P):
    # takes an object and a projection matrix (P) and projects the 3D
    # object orientation vector into the image plane.
    # compute rotational matrix around yaw axis
    R = np.array([[np.cos(obj.ry), 0, np.sin(obj.ry)],\
                 [0,              1, 0],\
                 [-np.sin(obj.ry), 0, np.cos(obj.ry)]])
    # orientation in object coordinate system
    orientation_3D = np.array([[0.0, obj.l],\
                               [0.0, 0.0],\
                               [0.0, 0.0]])
    # rotate and translate in camera coordinate system, project in image
    orientation_3D = np.dot(R,orientation_3D)
    orientation_3D[0] = orientation_3D[0] + obj.t[0]
    orientation_3D[1] = orientation_3D[1] + obj.t[1]
    orientation_3D[2] = orientation_3D[2] + obj.t[2]
    # vector behind image plane?
    if np.nonzero(orientation_3D[2]<0.1)[0].size != 0 :
        orientation_2D = [];
        return;
    # project orientation into the image plane
    orientation_2D = projectToImage(orientation_3D,P)
    return orientation_2D
```

```python
def drawBox2D(obj,img,P):
    # set styles for occlusion
    occ_col = ((49,205,49),(0,215,255),(0,0,255),(255,255,255))
    points = [int(x) for x in obj.box2D]
    cv2.rectangle(img, tuple(points[:2]), tuple(points[-2:]), \
        occ_col[obj.occlusion],2)
    if obj.truncation :
        cv2.rectangle(img, tuple(points[:2]), tuple(points[-2:]), \
        (0,0,0),1)
    if obj.type.lower() != 'dontcare':
        orient = computeOrientation3D(obj,P)
        orient = np.array(np.round(orient),dtype=int)
        cv2.line(img,tuple(orient[:,0]),tuple(orient[:,1]),(255,255,0),2)
        cv2.circle(img,tuple(orient[:,0]),1,(255,0,0),2)
def drawTrainData(img_idx):
    img1 = readImg_cv(image_dir,img_idx)
    img2 = img1.copy()
    P = readCalibration(calib_dir,img_idx,cam).reshape(3,4)
    objects = readLabels(label_dir,img_idx)

    for obj in objects:
        drawBox2D(obj,img1,P)
        
    img = np.concatenate((img1,img2),axis = 0)
    cv2.putText(img,'img_idx:%d' % img_idx,(30,30),\
                    cv2.FONT_HERSHEY_SIMPLEX,1,(255,)*3,2)
    return img
```

```python
def readImg_cv(img_dir,img_idx):
    return cv2.imread('%s/%06d.png' % (img_dir,img_idx))

import Image
def readImg(img_dir,img_idx):
    return Image.open('%s/%06d.png' % (img_dir,img_idx))
```

```python
save_dir = root_dir + data_set
def saveObj(type,img_idx):
    img = readImg(image_dir,img_idx)
    P = readCalibration(calib_dir,img_idx,cam)
    objects = readLabels(label_dir,img_idx);
    i = 0;
    for obj in objects:
        if (obj.type.lower() == type) and (obj.occlusion < 2):
            box = [int(x) for x in obj.box2D]
            roi = img.crop(box)
            path = save_dir+'/'+type.lower()+'/'
            if (obj.truncation < 0.5):
                path = path +'notTrunc/%s' % obj.objID +'.png'
                roi.save(path)
        i = i+1
```

```python
def countInOneImage(img_idx):
    P = readCalibration(calib_dir,img_idx,cam)
    objects = readLabels(label_dir,img_idx)
    return len(objects)
```

```python
#write to csv
import math
import csv
with open('kitti.csv','wb') as csvfile:
    writer = csv.writer(csvfile,delimiter = ',',quotechar='"', quoting=csv.QUOTE_MINIMAL)
    writer.writerow(('objID','type','occlusion','theta','ratio','w','h','alpha','truncation'))
    for img_idx in range(nImages):  
        img = readImg(image_dir,img_idx)
        P = readCalibration(calib_dir,img_idx,cam).reshape(3,4)
        objects = readLabels(label_dir,img_idx)
        for obj in objects:
            if (obj.type.lower() != 'dontcare') and (obj.occlusion < 2):
                w = obj.box2D[2]-obj.box2D[0]
                h = obj.box2D[3]-obj.box2D[1]
                ratio = w/h
                orient = computeOrientation3D(obj,P)
                if orient == None:
                    writer.writerow((obj.objID,obj.type,obj.occlusion,'',ratio,w,h,obj.alpha,obj.truncation))
                else :
                    xy = orient[:,1]-orient[:,0]
                    r = np.sqrt(sum(xy**2))
                    theta = math.atan2(-xy[1], xy[0]) 
                    writer.writerow((obj.objID,obj.type,obj.occlusion,theta,ratio,w,h,obj.alpha,obj.truncation))
    
```
